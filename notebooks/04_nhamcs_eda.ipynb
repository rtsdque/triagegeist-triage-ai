```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', 100)

# Load NHAMCS data
nhamcs = pd.read_csv('../data/nhamcs_data_2018_22.csv')

print(f"Shape: {nhamcs.shape}")
print(f"\nColumns:")
print(nhamcs.columns.tolist())
```

```python
print("=== TARGET DISTRIBUTION ===")
print(nhamcs['target_triage_acuity'].value_counts().sort_index())
print("\nAs percentages:")
print(nhamcs['target_triage_acuity'].value_counts(normalize=True).sort_index().mul(100).round(1))

print("\n=== DEMOGRAPHIC COLUMNS ===")
for col in ['sex', 'race', 'insurance']:
    print(f"\n{col}:")
    print(nhamcs[col].value_counts())

print("\n=== MISSING VALUES ===")
missing = nhamcs.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
print(missing)
```

```python
print("=== VITALS SUMMARY BY ACUITY LEVEL ===")
vitals = ['sys_bp', 'heart_rate', 'resp_rate', 'spo2', 'temp', 'pain_score']
summary = nhamcs.groupby('target_triage_acuity')[vitals].mean().round(2)
print(summary.to_string())
```

```python
# Check which competition features exist in NHAMCS
competition_features = ['age', 'sex', 'sys_bp', 'dias_bp', 'heart_rate', 
                        'resp_rate', 'spo2', 'temp', 'pain_score']

print("=== FEATURE OVERLAP ===")
for f in competition_features:
    exists = f in nhamcs.columns
    status = "YES" if exists else "MISSING"
    print(f"{f}: {status}")

print(f"\nAll NHAMCS columns:")
for col in nhamcs.columns:
    print(f"  {col}")
```
